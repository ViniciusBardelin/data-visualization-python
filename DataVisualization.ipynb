```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('imigrantes_canada.csv')
```

```python
dados.info()
```

```python
# Configurando a coluna País como índice:

dados.set_index('País', inplace = True)
```

```python
dados.head()
```

```python
# Vamos criar uma nova variável do intervalo de anos:

anos = list(map(str, range(1980, 2014)))
```

```python
print(anos)
```

```python
# Vamos criar um objeto que contem os dados de imigração do brasil ao longo dos anos:

brasil = dados.loc['Brasil', anos]
```

```python
brasil
```

```python
# O objeto acima é uma series. Vamos agora criar um dataframe utilizando um dicionário: 

brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

#### Trabalhando com Subplot

```python
fig, axs = plt.subplots(1,2, figsize = (15,5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()


axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para o Canadá')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()

plt.show()
```

A linha de código `fig, axs = plt.subplots(1,2, figsize = (15,5))` cria uma figura com duas plotagens (subplots), uma ao lado da outra, e define o tamanho da figura como 15 unidades de largura por 5 unidades de altura. Além disso:

- `fig, axs = plt.subplots(1,2, figsize=(15,5))`:
    - `plt.subplots(1, 2, figsize=(15, 5))`: Esta parte cria uma nova figura (`fig`) e uma matriz de subplots (`axs`). O parâmetro `1` indica que queremos 1 linha de subplots, e o parâmetro `2` indica que queremos 2 subplots em cada linha, totalizando 2 subplots. O parâmetro `figsize=(15, 5)` define o tamanho da figura como 15 unidades de largura por 5 unidades de altura.

    - `fig, axs = ...`: Este padrão de atribuição descompacta o objeto retornado pela função `plt.subplots()` em duas variáveis separadas: `fig` (a figura) e `axs` (os subplots). A variável `fig` é a figura que contém todos os subplots, enquanto `axs` é uma matriz de objetos de eixo (ou "axes") que representam cada subplot individualmente.

Portanto, essa linha de código cria uma figura com dois subplots posicionados lado a lado, cada um com dimensões definidas por `figsize=(15, 5)`. Isso é útil para comparar visualmente dois conjuntos de dados ou gráficos relacionados.

```python
fig, axs = plt.subplots(2,2, figsize = (10,6))
fig.subplots_adjust(hspace = 0.5, wspace =0.3)

axs[0,0].plot(dados.loc[('Brasil', anos)]) 
axs[0,0].set_title('Brasil')

axs[0,1].plot(dados.loc[('Colômbia'), anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(dados.loc[('Argentina'), anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(dados.loc[('Peru'), anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrantes')

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,6))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3)
fig.suptitle('Imigração para o Canadá')

axs[0,0].plot(dados.loc[('Brasil', anos)]) 
axs[0,0].set_title('Brasil')

axs[0,1].plot(dados.loc[('Colômbia'), anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(dados.loc[('Argentina'), anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(dados.loc[('Peru'), anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

#### Alterando Cores

```python
america_sul = dados.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive',
         'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
# para adicionar cor basta usar o argumento color abaixo:
ax.bar(america_sul.index, america_sul['Total'], color = cores)
# abaixo, com `loc = 'left'` podemos colocar o título do gráfico mais para esquerda:
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize = 15)
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes')
# abaixo, estamos definindo como 12 os ticks do eixo X:
ax.xaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive',
         'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul.index, america_sul['Total'], color = cores) # vamos mudar de `ax.bar` para `ax.barh`:
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize = 15)
ax.set_xlabel('Número de Imigrantes') # como invertemos o gráfico, devemos inverter os títulos dos eixos.
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
# Vamos começar criando um novo dataframe ordenado pela coluna Total:
america_sul_sorted = america_sul.sort_values('Total', ascending = True)


fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores) # mudando `america_sul` para `america_sul_sorted`.
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize = 15)
ax.set_xlabel('Número de Imigrantes')
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
plt.show()
```

#### Adicionando Anotações

```python
fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores) # mudando `america_sul` para `america_sul_sorted`.
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize = 15)
ax.set_xlabel('Número de Imigrantes')
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)

for i,v in enumerate(america_sul_sorted['Total']):
    ax.text(v+20, i, str(v), color = 'black', fontsize = 10, ha = 'left', va = 'center')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores) # mudando `america_sul` para `america_sul_sorted`.
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize = 15)
ax.set_xlabel('Número de Imigrantes')
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)

for i,v in enumerate(america_sul_sorted['Total']):
    ax.text(v+20, i, str(v), color = 'black', fontsize = 10, ha = 'left', va = 'center')

# O ax.set_frame_on(False) vai adicionar a borda do gráfico:
ax.set_frame_on(False)

# Abaixo, tiramos o eixo X da visualização, já que temos os valores total de cada país:
ax.get_xaxis().set_visible(False)

# Abaixo, tiramos os ticks (tracinhos) do eixo Y usando:
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```

```python
# Vamos começar salvando um código em PNG. Antes disso, vamos alterar as bordas do gráfico abaixo, deixando somente as linhas dos eixos X e Y:

fig, ax = plt.subplots(figsize = (8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá', fontsize = 15)
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# nas duas próximas linhas abaixo vamos remover a borda do topo do gráfico e da direita do gráfico:

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# salvando a figura com o nome imigracao_brasil_canada; com `transparent = False` não vamos deixar o fundo desse PNG transparente;
# com dpi podemos melhorar a qualidade da imagem;

fig.savefig('imigracao_brasil_canada.png', transparent = False, dpi = 300, bbox_inches='tight')

plt.show()
```
